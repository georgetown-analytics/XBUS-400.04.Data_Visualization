# Object-Oriented API

## Figure & Axes

Matplotlib uses a hierarchy of objects to display all of its plotting elements in its final output. The Figure object is the top-level container of our visualization and it can contain one or more Axes objects.

The Axes represent an individual plot and it is the area where our data will be plotted. It is important not to confuse it with the word "axis", which refers to the x/y axis of a plot. Axes is not the plural of "axis"! 

Here, we have called the `figure()` method from the `pyplot` class, which returns a `figure` object, and we've explicitly set the `facecolor` parameter to help differentiate it from the Axes. Next, we've added a blank Axes to it by calling this object's `add_axes` method.

**Matplotlib Methods**:
- `matplotlib.figure` API [documentation](https://matplotlib.org/api/figure_api.html)  
- `matplotlib.pyplot.figure` [documentation](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.figure.html)  
- `matplotlib.figure.Figure.add_subplot` [documentation](https://matplotlib.org/api/_as_gen/matplotlib.figure.Figure.html?highlight=add_subplot#matplotlib.figure.Figure.add_subplot)

```python
# Instantiate an instance of the Figure class
fig = plt.figure(facecolor=(1,0,0,.1))

# Add a single Axes subplot to the Figure
ax = fig.add_subplot()

# Display the plot
plt.show()
```

In the previous example, we first instantiated the Figure and then added the Axes object to it. However, matplotlib also allows users to combine these steps using `plt.subplots()`, which creates a new Figure with a single Axes when it is called with no arguments. This function returns a two-item tuple containing the Figure and Axes objects, which we've unpacked here into the variables `fig` and `ax`, respectively.

Matplotlib `pyplot.subplots` [documentation](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplots.html)

```python
# Instantiate a Figure with a single Axes
fig, ax = plt.subplots(facecolor=(1,0,0,.1))

# Display the plot
plt.show()
```

## Plot the Data

Now, let's plot some data! By default, matplotlib can accept input data as a Python list, NumPy array, or a Pandas DataFrame, so our data needs to be in one of those formats. Here, we will generate two NumPy arrays of sample data and then see how we can plot our data. (Please note, that their lengths must match, or else we will get a `ValueError`.)

You also see here another way we can add a title to our plot using Axes `set_title()` method, along with its methods to label our x- and y-axis. When adding the x-axis and y-axis of our plot, we can also use the optional `size` parameter to control the size of the text. Here, I've only set the `size` parameter for the x-axis to show the difference.

*Matplotlib Methods*:
- `Axes.plot` [documentation](https://matplotlib.org/api/_as_gen/matplotlib.axes.Axes.plot.html)  
- `Axes.set_facecolor` [documentation](https://matplotlib.org/api/_as_gen/matplotlib.axes.Axes.set_facecolor.html)

```python
# Create example data to display
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)

# Instantiate a Figure with a single Axes
fig, ax = plt.subplots(facecolor=(1,0,0,.1))

# Add data to the Axes object
ax.plot(x, y)

# Set the Axes color
ax.set_facecolor('#e1ddbf')

# Add a title
ax.set_title('Plot a Single Dataset', size=18)

# Label the x-axis
ax.set_xlabel('Label the x-axis', size=16)

# Label the y-axis
ax.set_ylabel('Label the y-axis')
```

## Legends

Many of the charts that we are going to be creating will use a lot of different visual properties&mdash;such as color, shapes, or sizes&mdash; to represent different values in our data. We can add a legend to our plot to identify the data in our visualization by these different properties.

One way to do this with matplotlib is to first pass the name of the element to the `plot()` method's optional `label` parameter and then call the Axes `legend()` function.

Matplotlib [Legend Guide](https://matplotlib.org/tutorials/intermediate/legend_guide.html)  
Matplotlib `Axes.legend` [documentation](https://matplotlib.org/api/_as_gen/matplotlib.axes.Axes.legend.html)

```python
# Instantiate a Figure with a single Axes
fig, ax = plt.subplots()

# Add data to the Axes object
ax.plot(x, y, label='Blue data')

# Add second dataset to the Axes object
ax.plot(y, x, label='Orange data')

# Add a title
ax.set_title('Add a Legend', size=16)

# Add a legend
ax.legend();
```

There are several different ways we can customize our legends using matplotlib.
- Remove the box around the legend by setting `frameon=False`
- Use a rounded box by setting `fancybox=True`
- Add a shadow behind the legend by setting `shadow=True`
- Add a title by passing a string to `title`
- Adjust the vertical spacing around the labels by passing a numeric value to `labelspacing`
- Adjust the padding around the text with `borderpad`
- Use the `ncol` parameter to specify the number of columns in the legend
- Adjust the size of the text using the optional `fontsize` parameter. If you pass a numeric value to this keyword, the size will be the absolute font size in points. However, you can also pass on of the available strings&mdash; `"xx-small"`, `"x-small"`, `"small"`, `"medium"`, `"large"`, `"x-large"`, `"xx-large"`. The string values are relative to the current default font size.

To change the position of the legend, we can take advantage of the `legend()` function's optional `loc` parameter. Matplotlib gives us the option to pass in both numerical and text values to this parameter. By default, it is set to `0` or `"best"`.

| Value | Location String | Value | Location String |
| :---- | :-------------: | :---: | :-------------: |
| `1` | `'upper right'` | `6` | `'center left'` |
| `2` | `'upper left'` | `7` | `'center right'` |
| `3` | `'lower left'` | `8` | `'lower center'` |
| `4` | `'lower right'` | `9` | `'upper center'` |
| `5` | `'right'` | `10` | `'center'` |

```python
# Instantiate a Figure with a single Axes
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12,6))

# Add data to the Axes object
ax1.plot(x, y, label='Blue data')
ax1.plot(y, x, label='Orange data')

# Add second dataset to the Axes object
ax2.plot(x, y, label='Blue data')
ax2.plot(y, x, label='Orange data')

# Add a legend
ax1.legend(loc='upper right', shadow=True, fancybox=True, ncol=2, fontsize='large')
ax2.legend(loc='center right', borderpad=2, labelspacing=2, title='Hi, I am a legend title!', title_fontsize=12);
```

# Multiple Subplots

While an Axes object can only belong to a single Figure, a Figure can contain multiple Axes, and in those cases, we refer to the Axes objects as subplots. This can be very helpful since eventually if we add too much data to a single Axes our plot will be too busy, obscuring patterns rather than revealing them. In these cases, we will often want to add multiple Axes objects to our Figure.

In our first examples, we did not specify the number of rows or columns, and therefore both `fig.add_subplot()` and `plt.subplots()` will only create a single subplot or Axes by default. Here is how we could have explicitly instructed matplotlib to add a single subplot to our Figure.

The `111` in the first example is a short alias for `1,1,1`, so we are passing three different arguments to `add_subplot()` method. The first argument designates the number of rows (`1`), the second the number of columns (`1`), and the third is the plot number or index (`1`) of the subplot.

- `Axes.subplots` [documentation](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplots.html)  
- `Figure.add_axes` [documentation](https://matplotlib.org/api/_as_gen/matplotlib.figure.Figure.html?highlight=add_axes#matplotlib.figure.Figure.add_subplot)

```python
fig, ax = plt.subplots(facecolor=(1,0,0,.1))                              # First method
ax = fig.add_subplot(111)
```

By default when we call `plt.subplots()`, this method is passed `fig.add_subplot(1,1,1)`, which tells it to add one Axes object onto a 1 row by 1 column grid.

The `plt.subplots()` function was designed to make it more convenient to create a set of subplots in common layouts in a single call. It takes `nrows` and `ncols` as positional arguments. However, if we do not pass in these parameters, the function is passed with `fig.subplots(1,1,1)` by default.

```python
# fig, ax = plt.subplots(nrows=1, ncols=1, facecolor=(1,0,0,.1))          # Second method
fig, ax = plt.subplots(1, 1, facecolor=(1,0,0,.1))
```

## Stacking subplots in one direction

When specifying subplots, remember that the first argument represents the number of rows and the second the number of columns. Here, we've specified that on our Figure, we want to create two Axes objects along a single row. In the first example, I've explicitly set the `nrows` and `ncols` parameters; however, we can achieve the same results by omitting the parameter names: `fig, axes = plt.subplots(1,2)`

```python
# Instantiate a Figure and add two Axes objects on a 1x2 grid
fig, axes = plt.subplots(nrows=1, ncols=2, facecolor=(1,0,0,.1))
```

When we are stacking our subplots in only one direction, as we've done here, the returned `axes` object is a 1D NumPy array containing a list of the Axes objects. For example, here we've instantiated a Figure stacking the two independent Axes subplots side-by-side by passing the parameters `1`, `2` for one row and two columns.

```python
# Instantiate the Figure and two Axes objects
fig, axes = plt.subplots(1, 2, facecolor=(1,0,0,.1))

print('Number of Axes objects:', len(axes))

# Print the 1D NumPy array of Axes objects
print(axes)
```

Now, let's vertically stack our subplots by passing the parameters `2` and `1` for two rows and one column.

```python
# Instantiate the Figure and two Axes objects on a 2x1 grid
fig, axes = plt.subplots(nrows=2, ncols=1, facecolor=(1,0,0,.1))

print('Number of Axes objects:', len(axes))

# Print the 1D NumPy array of Axes objects
print(axes)
```

Notice when we printed our `axes` variable we could see that it was a 1D NumPy array containing two Axes objects. This means that we can access each Axes individually by slicing the `axes` object.

```python
# Instantiate the Figure and two Axes objects
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(9,6))

# Add data to the first Axes object (row 1, column 1)
axes[0].plot(x, y)

# Add a title to the first Axes object
axes[0].set_title('Vertically Stacked Subplots', size=18)

# Add data to the second Axes object (row 2, column 1)
axes[1].plot(x, y)

# Label each y-axis
axes[0].set_ylabel('Row 1, Column 1', size=12)
axes[1].set_ylabel('Row 2, Column 1', size=12);
```

**Unpack Axes Objects**

Another option if you are only creating only a small number of Axes is to unpack the Axes objects immediately into separate variables. This allows us to use `ax1` instead of the more verbose `axes[0]`.

```python
# Instantiate a Figure with two Axes objects
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, figsize=(9,6))

# Add data to the first Axes object (row 1, column 1)
ax1.plot(x, y)

# Add a title to the first Axes object
ax1.set_title('Vertically Stacked Subplots', size=18)

# Add data to the second Axes object (row 2, column 1)
ax2.plot(x, y)

# Label each y-axis
ax1.set_ylabel('Row 1, Column 1', size=12)
ax2.set_ylabel('Row 2, Column 1', size=12);
```

## Stacking subplots in two directions

When stacking subplots in two direction, the order of the Axes objects is started from left to right and from top to bottom within the grid. Here, we'll starting by instantiating our Figure and then add each subplot to a 2x2 grid. 

```python
# Instantiate a Figure
fig = plt.figure(facecolor=(1,0,0,.1), figsize=(9,6))

# Add an Axes object to row 1, column 1 (id 1)
ax1 = fig.add_subplot(221)

# Add an Axes object to row 1, column 2 (id 2)
ax2 = fig.add_subplot(222)

# Add an Axes object to row 2, column 1 (id 3)
ax3 = fig.add_subplot(223)

# Add an Axes object to row 2, column 2 (id 4)
ax4 = fig.add_subplot(224)

# Add text to the Axes objects
ax1.text(0.5, 0.5, 'I am ax1 (221)', size=16, va='center', ha='center')
ax2.text(0.5, 0.5, 'I am ax2 (222)', size=16, va='center', ha='center')
ax3.text(0.5, 0.5, 'I am ax3 (223)', size=16, va='center', ha='center')
ax4.text(0.5, 0.5, 'I am ax4 (224)', size=16, va='center', ha='center');
```

However, rather than first creating our Figure and then adding each Axes object separately, we can use `plt.subplots()` to create our 2x2 grid. 

Now, when stacking subplots in two directions, the returned `axes` object is a 2D NumPy array, so each Axes object can be accessed using `axes[x, y]`, i.e., the first Axes object would be accessed using `axes[0, 0]`.

```python
# Instantiate a Figure with four Axes objects on a 2x2 grid
fig, axes = plt.subplots(2, 2, facecolor=(1,0,0,.1))

# Add text to the Axes objects
axes[0, 0].text(0.5, 0.5, 'I am ax1 (221)', size=14, va='center', ha='center')
axes[0, 1].text(0.5, 0.5, 'I am ax2 (222)', size=14, va='center', ha='center')
axes[1, 0].text(0.5, 0.5, 'I am ax3 (223)', size=14, va='center', ha='center')
axes[1, 1].text(0.5, 0.5, 'I am ax4 (224)', size=14, va='center', ha='center');
```

## Tight Layout

Let's add `tight_layout()` to the previous example to adjust the padding and minimize the overlap between our subplots.

Matplotlib `tight_layout` [documentation](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.tight_layout.html)

```python
# Instantiate a Figure with four Axes objects on a 2x2 grid
fig, axes = plt.subplots(2, 2, facecolor=(1,0,0,.1))

# Add text to the Axes objects
axes[0, 0].text(0.5, 0.5, 'I am ax1 (221)', size=14, va='center', ha='center')
axes[0, 1].text(0.5, 0.5, 'I am ax2 (222)', size=14, va='center', ha='center')
axes[1, 0].text(0.5, 0.5, 'I am ax3 (223)', size=14, va='center', ha='center')
axes[1, 1].text(0.5, 0.5, 'I am ax4 (224)', size=14, va='center', ha='center')

# Automatically adjust the spacing between the subplots
plt.tight_layout();
```

Here, we'll expand upon the previous example by adding the sample data we created earlier.

```python
# Instantiate a Figure and array of subplots with 2 rows and 2 columns
fig, axes = plt.subplots(2, 2)

# Add data to the top left (0,0) Axes object
axes[0, 0].plot(x, y)
axes[0, 0].set_title('Axis [0, 0]', size=14)

# Add data to the top right Axes (0,1) object
axes[0, 1].plot(x, y, 'tab:orange')
axes[0, 1].set_title('Axis [0, 1]', size=14)

# Add data to the bottom left (1,0) Axes object
axes[1, 0].plot(x, y, 'tab:green')
axes[1, 0].set_title('Axis [1, 0]', size=14)

# Add data to the bottom right (1,1) Axes object
axes[1, 1].plot(x, y, 'tab:red')
axes[1, 1].set_title('Axis [1, 1]', size=14)

# Automatically adjust the spacing between the subplots
plt.tight_layout();
```

An additional option is to use tuple unpacking to assign all of the subplots to their own dedicated variables.

Here, we also used matplotlib's `get_axes()` and `label_outer()` functions to hide the x-labels for the plots in the first row and to only add y-labels to the plots in the first column.

Matplotlib `Figure.get_axes` [documentation](https://matplotlib.org/api/_as_gen/matplotlib.figure.Figure.html?highlight=get_axes#matplotlib.figure.Figure.get_axes)  
Matplotlib `label_outer` [documentation](https://matplotlib.org/api/_as_gen/matplotlib.axes.SubplotBase.html?highlight=label_outer#matplotlib.axes.SubplotBase.label_outer)

```python
# Instantiate a Figure and array of subplots with 2 rows and 2 columns
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)

# Add data to the first Axes object (row 1, column 1)
ax1.plot(x, y)

# Add data to the second Axes object (row 1, column 2)
ax2.plot(x, y, 'tab:orange')

# Add data to the third Axes object (row 2, column 1)
ax3.plot(x, y, 'tab:green')

# Add data to the fourth Axes object (row 2, column 2)
ax4.plot(x, y, 'tab:red')

# Add x-labels and y-labels to each subplot 
for ax in fig.get_axes():
    ax.set(xlabel='x-label', ylabel='y-label')
    
# Hide x-labels and ticklabels for top plots and y-ticks for right plots
for ax in fig.get_axes():
    ax.label_outer()

# Automatically adjust the spacing between the subplots
plt.tight_layout();
```

## Sharing Axes

By default, each Axes object is scaled individually, so if the ranges are different, the tick values of the subplots will not align, as we can see in this example.

```python
# Create a Figure and a 1x2 grid (1 row and 2 columns)
fig, axes = plt.subplots(1,2, figsize=(10,2))

# Add data to the first Axes object (row 1, column 1)
axes[0].plot(x, y)

# Add data to the second Axes object (row 1, column 2)
axes[1].plot(x + 2, y + 2, 'tab:orange');
```

However, when calling `plt.subplots()` we can set the `sharex` or `sharey` parameters to `True` to align the horizontal or vertical axis. Let's try this with our previous example by setting `sharex=True` so that our subplots share the same scale on the x-axis.

```python
# Create a Figure and an array of axes (1 row and 2 columns) with shared x-axis
fig, axes = plt.subplots(1, 2, sharex=True, figsize=(10,2))

# Add data to the first Axes object (row 1, column 1)
axes[0].plot(x, y)

# Add data to the second Axes object (row 1, column 2)
axes[1].plot(x + 2, y + 2, 'tab:orange');
```

Let's now set both `sharex` or `sharey` to `True` so that our subplots share the same scale on both axes.

```python
# Create a Figure and an array of axes (1 row and 2 columns) with shared x- and y-axis
fig, axes = plt.subplots(1,2, figsize=(10,2), sharex=True, sharey=True)

# Add data to the first Axes object (row 1, column 1)
axes[0].plot(x, y)

# Add data to the second Axes object (row 1, column 2)
axes[1].plot(x + 2, y + 2, 'tab:orange');
```

# Saving our plot

Eventually, we will want to save our plots to disk and matplotlib offers us several different file formats to choose from. Here, we can see all of our available options.

```python
# Find the available file types
fig.canvas.get_supported_filetypes()
```

We can then use the `fig.savefig()` function to save the current Figure. Two of this method's most useful parameters are:
- `transparent=True` will save our Figure with a transparent background if the specified format supports it. This is often useful for plots that will be displayed on top of a colored background, such as on a web page.
- `dpi` controls the resolution (pixels/dots per square inch) of our Figure. A Figure of `figsize=(width, height)` will have a `dpi` of `pixel_width, pixel_height = width*dpi, height*dpi`. Typically, a `dpi` of 100 is fine for web images but a `dpi=300` may be needed for images intended for a written report or PDF.

Matplotlib `Figure.savefig` [documentation](https://matplotlib.org/api/_as_gen/matplotlib.figure.Figure.html?highlight=savefig#matplotlib.figure.Figure.savefig)

```python
# Create a Figure with the default values
fig, _ = plt.subplots()

# Find the Figure dpi
print('Figure dpi: ', fig.get_dpi())

# Find the Figure size in inches
print('\nFigure size (width, height):', fig.get_size_inches())
print('Default Figure size:', mpl.rcParams['figure.figsize'])

# Get the Figure width and height (inches)
print('\nFigure width:', fig.get_figwidth())
print('Figure height:', fig.get_figheight())

# Get the Figure width and height (pixels)
print('\nFigure pixel width:', fig.get_figwidth() * fig.get_dpi())
print('Figure pixel height:', fig.get_figheight() * fig.get_dpi())
```

Here, let's further explore the relationship between a Figure's size (`figsize`) and its resolution (`dpi`). 

Notice each of the following Figures have a `figsize` of `(2,2)`. However, we can see that by increasing the `dpi` it also increases the pixel size of our plot.

```python
# Read the image into an array
image = plt.imread('../images/john_hunter.jpeg')
image.shape
```

```python
# Save the starting Figure size and dpi
dpi = 72
figsize = (2,2)

def dpi_image_plot(fs, dpi):
    """Plot the image with varying figure and dpi sizes"""
    # Instantiate a Figure with a single Axes
    fig, ax = plt.subplots(figsize=fs, dpi=dpi)
    
    # Add a title
    ax.set_title('Figsize: {}, dpi: {}'.format(fs, dpi))
    print('\nFigure dpi:', fig.get_dpi())

    print('Figure size (inches):', fig.get_size_inches())
    print('Figure size (pixels):', fig.get_figwidth() * fig.get_dpi(), fig.get_figheight() * fig.get_dpi())
    
    # Display the image
    plt.imshow(image)
    
# Plot the same sized Figure at different DPI values
for i in range(1,4):
    dpi_image_plot(figsize, i*72)
```
