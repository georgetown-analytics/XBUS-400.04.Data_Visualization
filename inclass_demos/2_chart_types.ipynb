# Line Plot

Line graphs can be helpful when we want to detect trends in our data or when we want to understand changes to one variable as a function of time.

As we saw in our introduction to matplotlib, we can call `Axes.plot()` to plot our data, which by default will create a simple line plot.

Matplotlib `matplotlib.axes.Axes.plot` [documentation](https://matplotlib.org/api/_as_gen/matplotlib.axes.Axes.plot.html)

```python
# Create 2 NumPy arrays of sample data to display
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)

# Instantiate a Figure with a single Axes
fig, ax= plt.subplots()

# Add the data to the Axes object
ax.plot(x, y)

# Add a title
ax.set_title('Plot Two NumPy Arrays', size=18)

# Label the x- and y-axis
ax.set_xlabel("x-axis label", size=14)
ax.set_ylabel("y-axis label", size=14)

# Display the plot
plt.show();
```

Remember from our object-oriented walkthrough that Jupyter Notebooks will display the Figure without calling the  `plt.show()` command explicitly. However, if we were working in a Python shell or an embedded application, `plt.show()` or `plt.draw()` is required to display the Figure.

If only one argument is passed to the `plot()` method, it will take it as y-axis coordinates and use the element's index as the coordinates for the x-axis.

```python
# Instantiate a Figure with a single Axes
fig, ax = plt.subplots()

# Plot the data
ax.plot([2, 3, 4])

# Add a title
ax.set_title('Plot One NumPy Array', size=18);
```

Another common way to modify your line charts to change the line style from solid to dashed or dotted using the `linestyle` argument. You can pass the line style name as a string, such as `'dashed'` or use its corresponding shortcode, `'--'`.

| Line Style  | Description  |
|---|---|

| `'--'` or `'dashed'` | dashed line |
| `'-.'` or `'dashdot'` | dash-dotted line |
| `':'` or `'dotted'` | dotted line |
| `'None'` | draws nothing |
| `''` | draws nothing |

```python
# Instantiate a Figure with two Axes on one row
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(10,5))

# Plot the data with a dotted line
ax1.plot(x, color='pink', linewidth=5, linestyle=':')
ax1.set_title('Dotted Line Style', size=18)

# Plot the data with a dashed line
ax2.plot(x, color='pink', linewidth=5, linestyle='--')
ax2.set_title('Dashed Line Style', size=18);
```

# Pyplot 

Now that we've seen how we can create and customize our line plots using matplotlib's object-oriented interface, let's compare this to using the pyplot API.

The `plt.plot()` function automatically creates our Figure and Axes and it also plots a line chart by default. It expects two arguments, `x` and `y`, such as Python lists, NumPy arrays or Pandas DataFrames, for the x and y-axis respectively.

  `plt.plot([x], y, [fmt])`
    
The brackets around the `x` and `[fmt]` parameters are meant to show that they are optional. This means that if we only pass the function one argument it will be plotted along the y-axis. Here, we'll pass in only a single list containing three numbers `[2, 3, 4]`, so matplotlib will use the list's index for the x-axis and plot a line chart connecting the points `(0, 2)`, `(1, 3)`, and `(2, 4)`.

Also, please notice in the following example that we didn't need to explicitly instantiate our Figure and Axes. In addition, when we added elements to the plot, such as the title and the axis labels, we didn't need to reference the specific Axes to add them to since the `pyplot` module is state-based, which means that matplotlib automatically keeps track of our current Figure and Axes.

**Matplotlib Methods**  
`pyplot.plot` [documentation](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html)  
`pyplot.title` [documentation](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.title.html)   
`pyplot.xlabel` [documentation](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.xlabel.html)  
`pyplot.ylabel` [documentation](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.ylabel.html)  

```python
# Plot a single list
plt.plot([2, 3, 4])

# Add a title
plt.title('Line Plot (pyplot)', size=18)

# Label the x- and y-axis
plt.xlabel('x-axis label', size=14)
plt.ylabel('y-axis label', size=14);
```

To plot `x` versus `y`, here we've passed in two lists. However, since `x` is optional and the default values are `[0, , N-1]`, it results in the same plot.

```python
# Plot two lists
plt.plot([0, 1, 2], [2, 3, 4])

# Add a title
plt.title('Plot Two Lists', size=18)

# Label the x- and y-axis
plt.xlabel('x-axis label', size=14)
plt.ylabel('y-axis label', size=14);
```

Now, what if we want to plot two different lists in the same plot using matplotlib's pyplot API? We simply have to call `plt.plot()` again.

```python
# Plot the data
plt.plot([2, 3, 4])

# Plot the data
plt.plot([3.5, 2.5])

# Add a title
plt.title('Plot Two Lists', size=18);
```

### Format String

`plt.plot([x], y, [fmt])`
   
The `plt.plot()` method also an optional argument, the format string (`[fmt]`), which can be used to set the marker style, line style, and the color of the plot.

The format string has three optional parts, with the first part controlling the marker style, which is the symbol that is plotted at each data point; the second option controls the style of line that connects the data points and the third controls the color of the plot.

`fmt = '[marker][line][color]'`

The letters and symbols can be passed to the function as a string. For example, the default string is `'-b'`, which concatenates the color string, `'b'` for blue and with a line style string, `'-'`, which creates a solid blue line.

Since each of the parts is optional, if a value for one of the parts is not provided, then the value from the style cycle is used. In addition, this means if we only specify the marker style but not the line style, a line won't be drawn between our adjacent data points. This is useful for plotting discrete data where no interpolation between points is necessary.

Here are some examples you might try later on your own.

| Format String  | Description  |
|---|---|
| `'or'` | Red circles |
| `'-g'` | A green solid line |
| `'--'` | A dashed line with default color |
| `'^k'` | Black triangle_up markers connect by a dotted line |

Here, let's create a line chart using a dotted red line, so we will pass in the string `'r:'`.

```python
# Plot a dotted red line
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'r:')

# Adjust the x- and y-axis limits
plt.axis([0, 6, 0, 20]);
```

Matplotlib `pyplot.axis` [documentation](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.axis.html)

# Pandas

Since pandas is built on top of matplotlib, the plotting is handled interally by matplotlib, which can accessed via the pandas Series or DataFrame `plot()` attribute. In fact, `plot()` is a wrapper for `pyplot.plot()`, so we could use this method to create identical graphs to the ones we just produced using matplotlib.

By default, pandas `.plot()` command will create a line chart with each numeric column along the y-axis. If we want to create a different type of chart, such a bar chart or a pie chart, pandas offers two different available options. For the first option, we can pass the name of the plot type as a string to the `plot()` attribute's `kind` parameter. For example, the following would create a histogram using the values from the specified column in our DataFrame:  `df.column.plot(kind='hist')`

Here are some of the other most common plot types that you'll find yourself using as you first start to explore your data. 

| Chart Type | Description  |
|---|---|
| `'bar'` | Bar chart|
| `'barh'` | Horizontal bar chart|
| `'hist'` | Histogram |
| `'box'` | Box-and-whiskers plot |
| `'kde'` or `'density'` | Density plot |
| `'area'` | Area plots |
| `'scatter'` | Scatter plots |
| `'hexbin'` | Hexagonal bin plots| 
| `'pie'` | Pie plots |

In the second option, we can create these plots by treating `plot` as a field and then calling the plotting methods on the DataFrame or Series, such as `df.plot.hist()` or `df.plot.bar()`. One advantage of this method is that by typing `df.plot.` and then hitting the `<TAB>` key, you can quickly see all of the available plot methods. 

Pandas `DataFrame.plot` [documentation](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.plot.html)  
Pandas `Series.plot` [documentation](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.plot.html)

The pandas plotting methods all come with an `ax` parameter, which allows us to plot our data onto a specific Axes in our Figure. In addition, we can pass any additional arguments to the specific matplotlib plotting method. These arguments are collected by the plot method's `**kwargs` parameter and then passed to the correct underlying matplotlib functions.

Now, let's see how it works! For these examples, we'll start by using the `tips` dataset that is one of the example datasets that comes with the seaborn library that we will learn more about next week.

```python
# Load in the Titanic dataset
titanic = sns.load_dataset('titanic')
titanic.head()
```

If you call `df.plot()` without any arguments, it will plot the DataFrame's index against each numeric column in your dataset. Notice that we didn't need to instantiate our Figure or Axes objects, and this is because pandas `plot()` method is a wrapper for matplotlib's `pyplot.plot()`.

Also notice how pandas automatically created a legend for us by using the column names! Legends are really useful for distinguishing between multiple datasets.

# Bar Chart

Bar charts are another common data visualization type that are especially effective when you have data that can be split into multiple, discrete categories. In a bar chart, categorical data is representated as rectangular bars whose lengths are proportional to the values that they represent. 

Both matplotlib's `pyplot.bar(x, height)` and `Axes.bar(x, height)` methods work best with a simple sequence of x, y values, passed in as a list or NumPy array.

Matplotlib `pyplot.bar` [documentation](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.bar.html)  

```python
# Create a list of values to plot
data = [5, 25, 50, 20]

# Find the length of the list or NumPy array
idx_length = range(len(data))

# Create a simple bar chart
plt.bar(x=idx_length, height=data)

# Add a title
plt.title('Matplotlib Bar Chart (pyplot)', size=18)

# Label the x- and y-axis
plt.xlabel('Length of the list', size=12)
plt.ylabel('Data values', size=12);
```

We can show the exact same chart horizontally using `plt.barh(y, width)`. However, notice now that the length of the list is still passed in first, but to the `y` parameter and `data` is passed to `width`. In addition, here we'll explicitly set the color of each bar and it's corresponding y-tick label.

Matplotlib `pyplot.barh` [documentation](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.barh.html)  
Matplotlib `pyplot.yticks` [documentation](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.yticks.html)  

```python
# Create a list of colors to plot
colors = ['red', 'orange', 'green', 'blue']

# Create a simple bar chart
plt.barh(y=idx_length, width=data, color=colors)

# Add a title
plt.title('Horizontal Bar Chart (pyplot)', size=16)

# Label the x- and y-axis
plt.xlabel('Data values (width)', size=12)
plt.ylabel('Length of the list (y)', size=12)

# Explicitly label the y-ticks
plt.yticks(idx_length, colors);
```

Here, we'll recreate the first bar chart using matplotlib's object-oriented API.

Matplotlib `Axes.bar` [documentation](https://matplotlib.org/api/_as_gen/matplotlib.axes.Axes.bar.html)

```python
# Instantiate a Figure and a single Axes
fig, ax = plt.subplots()

# Use matplotlib to create a bar chart
ax.bar(x=idx_length, height=data, color='orange')

# Add a title
ax.set_title('Matplotlib Bar Chart (OOP)', size=18)

# Label the x- and y-axis
ax.set_xlabel('Length of the list', size=12)
ax.set_ylabel('Data values', size=12);
```

## Horizontal Bar Chart

And here is how we would recreate the horizontal bar chart. However, this time let's pass in a list to the `color` parameter to explicitly set the color for each bar and a list to the `label` parameter to name each bar.

Please note, that when using `Axes.set_yticklabels()` to label your tick marks, this method should only be used after fixing the tick positions using `Axes.set_yticks`. Otherwise, the labels may end up in unexpected positions.rw

Matplotlib `Axes.barh` [documentation](https://matplotlib.org/api/_as_gen/matplotlib.axes.Axes.barh.html)  
Matplotlib `Axes.set_yticks` [documentation](https://matplotlib.org/api/_as_gen/matplotlib.axes.Axes.set_yticks.html)  
Matplotlib `Axes.set_yticklabels` [documentation](https://matplotlib.org/api/_as_gen/matplotlib.axes.Axes.set_yticklabels.html)

```python
# Instantiate a Figure and a single Axes
fig, ax = plt.subplots()

# Use matplotlib to create a horizontal bar chart
ax.barh(y=idx_length, width=data, color=['red', 'orange', 'green', 'blue'])

# Add a title
ax.set_title('Horizontal Bar Chart (OOP)', size=18)

# Label the x- and y-axis
ax.set_xlabel('Data values', size=12)
ax.set_ylabel('Length of the list', size=12)

# Set the y-axis tick locations
ax.set_yticks(idx_length)

# Set the y-axis labels with list of string labels
ax.set_yticklabels(colors);
```

Since most often our data is stored in pandas DataFrames instead of NumPy arrays, let's look at how we can take advantage of the fact that the pandas library is built on top of matplotlib.

By now, you should be familiar with pandas `pandas.Series.value_counts()` method which returns a Series containing counts of the unique values in a column. Here, let's use it find the number of men and women in the dataset. 

Pandas `Series.value_counts` [documentation](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.value_counts.html)

```python
# Create a bar chart using pandas
tips['sex'].value_counts().plot(kind='bar')

# Add a title
plt.title('Pandas Bar Chart', size=18);
```

Another convenient option is to chain the `value_counts()`, `plot()`, and `bar()` methods together.

Pandas `Series.bar` [documentation](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.plot.bar.html)

Use pandas `Series.value_counts()` method with an argument `normalize` set to `True` to find the relative frequencies of the unique values. Let's also set `rot=0` to make x-axis labels easier to read.

```python
# Create a normalized bar chart using pandas
tips['sex'].value_counts(normalize=True).plot.bar(rot=0)

# Add a title
plt.title('Pandas Bar Chart (normalized)', size=16);
```

We use pandas `barh()` function to generate a horizontal bar chart of our data. 

Pandas `Series.barh` [documentation](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.plot.barh.html)

```python
# Create a bar chart using pandas
tips['day'].value_counts().plot.barh()

# Add a title
plt.title('Pandas Horizontal Bar Chart', size=16)

# Label the x- and y-axis
plt.xlabel('Total bill ($)', size=12)
plt.ylabel('Day of the week', size=12);
```

Now, what if we want to visualize the total bill by the day of the week? We can combined pandas `groupby()` method with the `plot()` and `barh()` methods.

```python
tips[['day', 'total_bill']].groupby('day').sum().plot.barh(color='orange', legend='')

# Add a title
plt.title('Daily Total Sales', size=16)

# Label the x- and y-axis
plt.xlabel('Total bill ($)', size=12)
plt.ylabel('Day of the week', size=12);
```

## Stacked Bar Chart

Stacked bar plots share a lot of the strengths and weaknesses of univariate bar charts. They tend to work best for nominal categorical or small ordinal categorical variables.

Here, to create a stacked bar chart we can use pandas `groupby()` method. First, we group the data by the `"sex"` and `"smoker"` columns, before calling the `size()`, `unstack()` and `plot()` methods. This let's us see the number of male and female customers, split into smoker and non-smokers.

Pandas `DataFrame.groupby` [documentation](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.groupby.html)  
Pandas `DataFrame.size` [documentation](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.size.html)  
Pandas `DataFrame.unstack` [documentation](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.unstack.html)

```python
# Groupby the "sex" and "smoker" columns
ax = tips.groupby(['sex', 'smoker']).size().unstack().plot(kind='bar', stacked=True, rot=0)

# Add a title
ax.set_title('Pandas Stacked Bar Chart', size=16)

# Label the x-axis
ax.set_xlabel('Sex', size=12)

# Add a legend
ax.legend(['Smoker', 'Non-Smoker']);
```

# Scatter Plot

A scatter plot visualize the joint distribution of two variables where each point represents an observation from the dataset, allowing us to see if there is a meaningful relationship between them. In a scatter plot, we can see what happens to the dependent variable (y) when we change the independent variable (x).

As we learned earlier, we can format the line style of our plot by passing in the format string to matplotlib's `plot()` method. By default, it creates a solid blue line (`'-b'`); however, by using replacing the line style `-` with `.`, we can turn our graph into a scatter plot.

If we passed in the format string `'go-'`, matplotlib would create a green-colored dots with a solid line connecting them. However, by only passing in either `'go'` or `'g.'`, we can create a scatter plot with either dots or respectively. In other words, a scatter plot!

```python
# Instantiate a Figure with a single Axes
fig, ax = plt.subplots()

# Plot the data
ax.plot(tips['total_bill'], tips['tip'], 'go')

# Add a title
ax.set_title('Matplotlib plot() Method\nScatter Plot', size=16)

# Label the x- and y-axis
ax.set_xlabel('Total bill ($)', size=12)
ax.set_ylabel('Tip ($)', size=12);
```

In the previous example, it looks like there is a positive correlation between a `"total_bill"` and `"tip"` columns. This means that as the total cost of the bill increases, so does the amount that the customer tipped.

We can use pandas `groupby()` method to color our points by a third, categorical variable. This will help us identify trends between the sub-groups in our data.

In addition, here we have added a legend to our table by calling `plt.legend()`. Legends are very useful for helping viewers distinguish between data sources being displayed in our plot.

Matplotlib `pyplot.legend` [documentation](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.legend.html)

```python
# Group by a categorical variable
groups = tips.groupby('sex')
for name, group in groups:
    plt.plot(group['total_bill'], group['tip'], marker='o', linestyle='', label=name)

# Add a title
plt.title('Group by a Categorical Variable', size=16)

# Label the x- and y-axis
plt.xlabel('Total bill ($)', size=12)
plt.ylabel('Tip ($)', size=12)

# Add a legend
plt.legend();
```

## Marker Size

In our last plot, it was difficult to see the relationship in the &#0036;10-30 total bill range. However, we can fix this by changing the size of our markers with the `s` argument. It is defined in the documentation as:

> `The marker size in points**2. Default is rcParams['lines.markersize'] ** 2.`

This means that `s` parameter differs from the other `size` parameters in matplotlib's plotting methods since the marker size is in points squared (`points ** 2`) and the default is generated by squaring the `lines.markersize` value. So for example, if we wanted the markers to be 10 points high, then we would set the parameter as `s=100` or `s=10**2`.

Since the default `markersize` is 6, this sets the size of our markers to `36`. And since in the previous plot it was difficult to see the relationship in the \\$10 to \\$30 dollar range for the total bill, let's adjust the marker size to help us see this better. Here, let's reduce the marker size so that it is easier to see the relationship between the two variables.

```python
mpl.rcParams['lines.markersize']
```

```python
# Plot the data
plt.scatter(tips['total_bill'], tips['tip'], s=20)

# Add a title
plt.title('Scatter Plot\nAdjust the marker size [pyplot]', size=16)

# Label the x- and y-axis
plt.xlabel('Total bill ($)', size=12)
plt.ylabel('Tip ($)', size=12);
```

Another way to add information about a third variable to our scatter plot is by passing an array to the `s` parameter. This let's us set the size of each marker individually.

Here, we will do this by passing in the dataset's `'size'` column to `s`. Remember that the `'size'` column contains the total number of people in the party and ranges from 1 to 6. Now we can see that not only does the tip increase when total bill increases, but serving more people leads to a bigger tip as well.

*Please note, here I've squared each value and multipied it by 3 in order to make size difference more pronounced in the plot.*

```python
# Save the values from the "size" column as a NumPy array
size_of_table = tips['size'].to_numpy()

# Scale the marker size
size_of_table_scaled = [3 * s ** 2 for s in size_of_table]

# Create a scatter plot with the markers scaled by the 'size' column
plt.scatter(tips['total_bill'], tips['tip'], s=size_of_table_scaled)

# Label the x- and y-axis
plt.xlabel('Total bill ($)', size=12)
plt.ylabel('Tip ($)', size=12);
```

# Histograms

Histograms are used to visualize the statistical distribution of a continuous variable. The data points are split into even, discretely-spaced bins and the height of the bars indicates the number of data points in each bin.

Python offers several different options for plotting and customizing histograms. Here, we'll start by using the `pyplot.hist()` function from matplotlib's `pyplot` module, which is an interface to the `Axes.hist()` method that we'll look at next.

Matplotlib `pyplot.hist()` [documentation](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.hist.html)

```python
# Create a histogram with matplotlib's pyplot module
plt.hist(tips['total_bill'])

# Add a title
plt.title('Histogram (pyplot)', size=16)

# Label the x- and y-axis
plt.xlabel('Total bill ($)', size=12)
plt.ylabel('Frequency', size=12)

# Add grid lines
plt.grid(axis='y');
```

Here, we'll recreate the previous histogram using matplotlib's object-oriented API. However, let's also set `edgecolor='k'` to draw black lines around our bars. Also, notice here that we've used a different method to add the horizontal grid lines to our chart.

Matplotlib `Axes.hist()` [documentation](https://matplotlib.org/api/_as_gen/matplotlib.axes.Axes.hist.html)

```python
# Instantiate a Figure with a single Axes
fig, ax = plt.subplots()

# Plot a histogram using matplotlib
ax.hist(tips['total_bill'], edgecolor='k')

# Add a title
ax.set_title('Histogram (OOP)', size=18)

# Label the x- and y-axis
ax.set_xlabel('Total bill ($)', size=12)
ax.set_ylabel('Frequency', size=12)

# Add grid lines
ax.yaxis.grid();
```

The most important argument to consider when plotting a histogram is the number of bins the data should be grouped into. For both methods, the `bins` parameter to set to `10` by default, but it is important to experiment with different bin sizes.

Here, we can see the impact by plotting our data using 5, 10, 15, and 20 bins.

```python
# Instantiate the Figure and four Axes objects
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12,6))

# Add a title to the Figure
fig.suptitle('Histograms Bin Size Comparison', fontsize=22);

# Plot the data using only 5 bins
axes[0, 0].hist(tips['total_bill'], bins=5, edgecolor='k')
axes[0, 0].set_title('5 Bins', size=18)

# Plot the data using the default (10) number of bins
axes[0, 1].hist(tips['total_bill'], edgecolor='k')
axes[0, 1].set_title('10 Bins (default)', size=18)

# Plot the data using 15 bins
axes[1, 0].hist(tips['total_bill'], bins=15, edgecolor='k')
axes[1, 0].set_title('15 Bins', size=18)

# Plot the data using 20 bins
axes[1, 1].hist(tips['total_bill'], bins=20, edgecolor='k')
axes[1, 1].set_title('20 Bins', size=18)

# Adjust the layout
plt.tight_layout();
```

Here, let's look at another way to compare the impact of adjusting the histogram's `bins` argument by calling `plt.hist()` four times, which will plot all of the histograms together in one plot.

We'll also use the `alpha` parameter to decrease the transparency of the bars to help us see the different distributions better (`0.0` is transparent and `1.0` is opaque).

```python
# Plot the data with 4 different bin sizes
plt.hist(tips['total_bill'], bins=5, label='5 bins', alpha=0.7, edgecolor='tab:gray')
plt.hist(tips['total_bill'], label='10 bins (default)', alpha=0.7, edgecolor='tab:gray')
plt.hist(tips['total_bill'], bins=15, label='15 bins', alpha=0.7, edgecolor='tab:gray')
plt.hist(tips['total_bill'], bins=20, label='20 bins', alpha=0.7, edgecolor='tab:gray')

# Add a title and label the x-axis
plt.title('Histograms Bin Size Comparison', size=16)
plt.xlabel('Total bill ($)', size=12)

# Add a legend
plt.legend();
```

Pandas `Series.plot.hist()` uses `matplotlib.pyplot.hist()` to generate a histogram of our data. When using pandas' `hist()` method there is no need for the initial `groupby()` function like we did creating our bar plots since `hist()` automatically groups our data into bins.

# Box Plots

Similar to histograms, box plots help give us a sense of the central tendency and the spread of our data.

In the center of the box, the horizontal orange line represents the data's median value. When the distribution of the data is symmetric, the median line will be in the exact center of the box.

The lower-horizontal edge of the box specifies the first quartile (Q1) of the dataset, while the upper-horizontal edge specifies its third quartile (Q3). The difference between the Q1 and Q3 quartiles is known as the Inter-Quartile Range (IQR) and it covers the center of the distribution and contains 50% of the observations.

The upper whisker represents the higher values in your dataset and can be calculated with the following formula: `Q3 + IQR(1.5)`. The lower whisker represents lower values in your dataset and can be found: `Q1 - IQR*1.5)`. Data points that fall outside of the range of the whiskers are considered outliers and are plotted as circles. If the data contains outliers and they are not evenly present on both sides of the box, our distribution will be skewed and the orange median line will be off to the opposite side as you can see here. Also, notice how this creates an imbalance in the whisker lengths, with the longer tail having significantly more outliers.

To begin, let's use matplotlib's pyplot module to create a box plot for a single numeric feature. However, it is important to keep in mind that we will get an error if we try to use this method to plot a column with missing values.

# Pie Charts

<img src="../images/pie.png" alt="Pie" style="width:342px;"/>

Pie charts are a circular statistical graphic, which is divided into slices or wedges to the illustrate numerical proportions in the dataset. They can be an effective way to represent information when we are primarily interested in comparing a wedge against the whole pie, instead of wedges against each other.

Matplotlib provides the `pie()` function to plot pie charts from an array `X`. However, we will also need to use pandas `value_counts()` method to generate the unique category totals. Matplotlib then creates the wedges proportionally, so that each value `x` of the array generates a wedge proportional to `x/sum(X)`.

By setting `autopct='%1.1f%%'` we can format the values and `startangle=90` sets the angle by which the start of the pie is rotated, counterclockwise from the x-axis. In addition, notice here that we've passed a dictionary to the optional `textprops` parameter to format the font size and color of the text.

The `equal` aspect ratio ensures that pie chart is drawn as a circle by changing the axis limits, since the default is an ellipse. In addition, by setting `explode=(0, 0.1, 0, 0)` we've plotted the second wedge apart.

Matplotlib `Axes.pie` [documentation](https://matplotlib.org/api/_as_gen/matplotlib.axes.Axes.pie.html)  
Matplotlib `Axes.axis` [documentation](https://matplotlib.org/api/_as_gen/matplotlib.axes.Axes.axis.html)

```python
tips['day'].count()
tips.day.value_counts(normalize=True)
```

```python
# Instantiate the Figure and Axes
fig, ax = plt.subplots(figsize=(5,5))

ax.pie(tips['day'].value_counts(),
       labels=tips['day'].unique().tolist(),
       autopct='%1.1f%%',
       explode = (0.1, 0, 0, 0),
       startangle=90,
       textprops={'fontsize': 12, 'color': 'white'})

# Add a title 
ax.set_title('Matplotlib Pie Chart', size=18)

# Add a legend and a title legend
ax.legend(bbox_to_anchor=(1, 0, 0.25, 0.8), fontsize=10, title='Day of the Week')

# Equal aspect ratio ensures that pie is drawn as a circle
ax.axis('equal');  
```

Pandas `pyplot.pie` [documentation](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.pie.html)  
Matplotlib `pyplot.axis` [documentation](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.axis.html?highlight=pyplot%20equal)

```python
# Plot a pie chart
plt.pie(tips['sex'].value_counts(),
        autopct='%1.1f%%',
        startangle=90,
        textprops={'fontsize': 12, 'color': 'white'},
        explode=(0, 0.1))

# Add a title 
plt.title('Matplotlib Pie Chart', size=16)

# Add a legend and a title legend
plt.legend(['Female', 'Male'], bbox_to_anchor=(1,0,0.08,0.8), fontsize=10, title='Sex')

# Set equal aspect ratio
plt.axis('equal');
```
